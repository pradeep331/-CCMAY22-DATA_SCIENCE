```python
#Preprocessing dataset
def remove_pattern(input_txt,pattern):
    r=re.findall(pattern,input_txt)
    for word in r:
        input_txt=re.sub(word,"",input_txt)
    return input_txt    
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
lb
```

```python
data.columns
```

```python
data['sentiment']= lb.fit_transform(data['sentiment'])
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bagwords_vectorizer=CountVectorizer(max_df=0.90,min_df=2,max_features=1000,stop_words="english")
```

```python
bagwords=bagwords_vectorizer.fit_transform(data["review"])
```

```python
bagwords[0].toarray()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(bagwords,data["sentiment"],test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
f1_score(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
